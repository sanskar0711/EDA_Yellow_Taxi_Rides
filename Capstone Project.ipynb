Importing library matplotlib so that we can show data in the form of graphs for better understanding 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(30,15),facecolor=(.6,.9,.4,1))
plt.style.use('seaborn-whitegrid')
ax1=fig.add_subplot(331)
plt.title('Vendor ID')
plt.xlabel('Vendors 1 & 2')
plt.ylabel('Number of Yellow Taxi')
ax2=fig.add_subplot(332)
vendor=yellowtaxi['VendorID'].value_counts().head()
numv=[2,1]
bar=ax1.bar(numv,vendor)
bar[0].set_color('pink')
bar[1].set_color('cyan')
print(vendor)
passenger=yellowtaxi['passenger_count'].value_counts().head()
passenger
nump=[1,2,3,5,6]
plt.title('Passengers')
barp=ax2.bar(nump,passenger)
barp[0].set_color('pink')
barp[1].set_color('cyan')
print(passenger)
plt.xlabel('No. Of Passenger')
```

VENDOR ID-Here in first graph we can see that number of yellow taxi that are from Vendor ID 1(Creative Mobile Technologies) are about 355306 and Vendor ID 2(VeriFone Inc) are about 693269 so from this we are able to analyse that there are 337963 yellow taxi more in Vendor ID2 as compared to the Vendor ID1.

PASSENGERS-Here in second graph we are able to analyse that more people travel alone maybe for work or to go anywhere as compared to other number of passengers that travel i.e two, three, five and six for some reason four people don't travel together as shown in the data that I have picked up. 

```python
fig=plt.figure(figsize=(20,10), facecolor=(.6,.9,.4,1))
ax1=fig.add_subplot(111)
distance=yellowtaxi['trip_distance'].value_counts().head()
numb=[0.8,0.9,1.0,0.7,1.1]
ax1.plot(numb,distance)
plt.title('Trip Distance')
print(distance)
plt.xlabel('Distance Traveled')
plt.ylabel('No. of Passengers')
```

DISTANCE-From the above graph we can analysis that people on an average basis travel for nearly o.8Km to 1.0Km

```python
fig=plt.figure(figsize=(20,10), facecolor=(.6,.9,.4,1))
ax1=fig.add_subplot(121)
flag=yellowtaxi['store_and_fwd_flag'].value_counts().head()
numf=['N','Y']
ax1.bar(numf,flag)
print(flag)
plt.title('Store and Forward Flag')
plt.xlabel('Yes or No')
plt.ylabel('No. of times')
rate=yellowtaxi['RatecodeID'].value_counts().head()
print(rate)
ax2=fig.add_subplot(122)
numr=[1,2,5,3,4]
ax2.bar(numr,rate)
plt.title('Rate Code ID')
plt.xlabel('Different Rate Codes')
plt.ylabel('No. of times')
```

STORE AND FORWARD FLAG- From the graph 1 we are able to analyse that it happens very rarely when the taxi has no connection with the server and the details are stored for sometime then forwarded later to the server when the taxi is in range.

RATE CODE ID- From the graph 2 we can say that mostly rate code 1(Standard rate) is implemented. Sometimes rate code 2(JFK) is implemented which is applied when someone take taxi from the JFK airport which is an international airport the reason that people don't take yellow taxi from the airport can be due to high fare rates, anyone comes to pick them up or maybe they prefer some other taxi other than the yellow one, there can be any other personal reason as well. Rate code 3(Newark) and 4(Nassau or Westchester) are applied very rarely and rate code 5(Negotiated fare) is applied sometimes. 

Let's see which Payment Type people usually prefer

```python
fig=plt.figure(figsize=(20,10), facecolor=(.6,.9,.4,1))
ax1=fig.add_subplot(211)
pay=yellowtaxi['payment_type'].value_counts().head()
print(pay)
nump=["Credit Card","Cash","No Charge","Dispute"]
ax1.barh(nump,pay)
plt.title('Payment Type')
plt.ylabel('Different Payment Codes')
plt.xlabel('No. of times')
ax2=fig.add_subplot(212)
fare=yellowtaxi['fare_amount'].value_counts().head()
print(fare)
numf=[6.0,6.5,5.5,7.0,7.5]
ax2.scatter(numf,fare)
ax2.plot(numf,fare)
plt.title('Fares')
plt.xlabel('Fares Paid')
plt.ylabel('No. of Passengers')
```

PAYMENT TYPE- After seeing graph 1 we can say that mostly people pay through credit card(i.e payment code 1) and also many people use to pay fare through cash(i.e payment code 2). Remaining Payment code 3 and 4 are very rarely used that is No Charge and Dispute.

FARES- Seeing graph 2 we are able to analyse that on an average the fare is nearly from 6-7.5 dollars roughly also depends upon the amount of tip passenger is providing.

```python
ex=yellowtaxi['extra'].value_counts().head()
print(ex)
tip=yellowtaxi['tip_amount'].value_counts().head()
print(tip)
fig=plt.figure(figsize=(20,10), facecolor=(.6,.9,.4,1))
ax1=fig.add_subplot(221)
nume=['0.0','2.5','0.5','1.0','3.0']
ax1.bar(nume,ex)
plt.title('Extra Amount Paid')
plt.xlabel('Extra Paid(in dollars)')
plt.ylabel('No. of Passengers')
ax2=fig.add_subplot(222)
numt=['0.00','1.00','2.00','1.50','2.06']
ax2.bar(numt,tip)
plt.title('Tip Paid by Passengers')
plt.xlabel('Tip Paid(in dollars)')
plt.ylabel('No. of Passengers')
fig=plt.figure(figsize=(20,10), facecolor=(.6,.9,.4,1))
ax3=fig.add_subplot(111)
ax3.plot(nume,numt)
plt.title('Relation Between Tips paid and Extras Charged')
plt.xlabel('Extras Paid(in dollars)')
plt.ylabel('Tip Paid(in dollars)')
```

EXTRAS CAHRGED- Graph 1 tells us that very less extras are charged that is most of time passengers don't have to pay any extra charges and if they have to pay in some cases like Miscellaneous extras and surcharges. Currently, this only includes
the 0.50 and 1 dollar for rush hour and overnight charges, maximum extra amount charged to any passenger was 3 dollars.

TIPS GIVEN- Graph 2 tells us that very less people give Tips to the Driver and the maximum amount given was nearly about 2.06 dollars, also in Newyork drivers don't take tips so usually because they prefer to earn money only through their jobs and not through any other medium like tips. 

RELATION BETWEEN TIPS PAID AND EXTRA CHARGES-From graph 3 we can observe that they both are directly proportional to each other that is very less people have to pay extra charges and also very less people pay tips to the drivers also we can relate that people who's payment mode is cash they only pay the tips while in the case of credit card no tips are paid because the exact total amount is deducted from the cards. 

MTA_TAX- From the CSV file that we have loaded in our dataframe, we can easily observe that a fixed amount of near about 0.5 Dollars is charged from almost all the passengers.
